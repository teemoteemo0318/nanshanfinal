# **理賠再購合併**

## 被(與被有關係的人)vs被 合併

```python
df3['再購(120天)'] = 0 
for index, item in enumerate(df3.index.get_level_values(0)):
    time_list = df4.query('INSURED_RK in @rk_dict[{}]'.format(item))['EFFECTIVE_DT'].unique()
    for time in time_list:
        if (df3.iloc[index,6] < time) and (time < df3.iloc[index,11]):
            df3.iloc[index,14] = 1
            break
    if index % 10000 == 0:
        print("第{}筆資料已標記!!".format(index))
```

```python
df3['再購(180天)'] = 0 
for index, item in enumerate(df3.index.get_level_values(0)):
    time_list = df4.query('INSURED_RK in @rk_dict[{}]'.format(item))['EFFECTIVE_DT'].unique()
    for time in time_list:
        if (df3.iloc[index,6] < time) and (time < df3.iloc[index,12]):
            df3.iloc[index,15] = 1
            break
    if index % 10000 == 0:
        print("第{}筆資料已標記!!".format(index))
```

```python
df3['再購(360天)'] = 0 
for index, item in enumerate(df3.index.get_level_values(0)):
    time_list = df4.query('INSURED_RK in @rk_dict[{}]'.format(item))['EFFECTIVE_DT'].unique()
    for time in time_list:
        if (df3.iloc[index,6] < time) and (time < df3.iloc[index,13]):
            df3.iloc[index,16] = 1
            break
    if index % 10000 == 0:
        print("第{}筆資料已標記!!".format(index))
```
