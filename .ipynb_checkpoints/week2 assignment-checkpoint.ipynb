```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv'
df = pd.read_csv(file)
df['Data_Value']=df['Data_Value']/10
df['year'] = df['Date'].apply(lambda x:x[:4])
df['day'] = df['Date'].apply(lambda x:x[5:])
```

```python
df1 = df[df['year']!='2015']
df1 = df1[df1['day']!='02-29']
df2 = df[df['year']=='2015']
```

```python
df1 = df1.groupby('day').Data_Value.agg(max=np.max,min=np.min)
df2 = df2.groupby('day').Data_Value.agg(max=np.max,min=np.min)
```

```python
date = list(df1.index)
```

```python
max2015 = list(df2['max'])
min2015 = list(df2['min'])
xmax = list(df1['max'])
xmin = list(df1['min'])
new_high = []
new_h_index = []
new_low = []
new_l_index = []
for i,value in enumerate(max2015):
    if value > xmax[i]:
        new_high.append(value)
        new_h_index.append(i)
for i,value in enumerate(min2015):
    if value < xmin[i]:
        min2015[i] = 0
        new_low.append(value)
        new_l_index.append(i)
```

```python
plt.figure(figsize=(9,6))
plt.plot(xmax,label='2005-2014 max')
plt.plot(xmin,label='2005-2014 min')
plt.fill_between(range(len(xmax)),xmax,xmin,alpha=0.25,facecolor='gray' )
plt.scatter(new_h_index,new_high,c='red',label='2015 new high')
plt.scatter(new_l_index,new_low,c='green',label='2015 new low')
plt.legend(frameon=False)
plt.title('Max and min Temperature of Day')
ax = plt.gca().xaxis

i = 1
for spine in plt.gca().spines.values():
    if i==2 or i ==4:
        spine.set_visible(False)
    i+=1
list1 = [0,31,59,90,120,151,181,212,243,273,304,334]
ax.set(ticks=list1, ticklabels=calendar.month_abbr[1:13])
plt.xlim(0, 365)
plt.ylabel('â„ƒ',rotation=0)

plt.savefig('week2.png')
```
